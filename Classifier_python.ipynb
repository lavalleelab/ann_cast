## Classification

- Features are taken from HCA reference data (which was run with Cell Ranger (version 5) and  GRCh38 standard reference (Ensembl 98) )

```python
# load features
features = pd.read_csv("Model/ANNClassifier_features.txt",header=None)
features = features[0].to_list()
# load the model 
model = pickle.load(open('Model/ANNClassifier.sav', 'rb'))
```

```python
def ANNpredictions_V2(adata,features= features, classifier_model= model):
        
    #adata was filtering for unexpressed genes. We add genes that are
    #in the model but not in adata. We add them as zero in a new adata
    absent_features = list(set(features) - set(adata.var.index)) 

    #Create a sparse matrix or zeros
    temp = sp.csr_matrix( (len(adata.obs.index),len(absent_features)), dtype=np.double)    
        
    #Create an index of column names, since csr_matrix dont have column names
    colnames = pd.Index(adata.var.index.to_list() +absent_features)
    
    #concatenate sparse matrices
    final_df = sp.hstack((sp.csr_matrix(adata.X), temp), format='csr') 
       
    #Reorder the columns in the expected order
    df_test_all_ct =final_df[:,colnames.get_indexer_for(features)] 

    # pred_others_cell_types = classifier_model.predict(df_test_all_ct)
    # adata.obs['ANNprediction_V1'] = pred_others_cell_types

    pred = classifier_model.predict_proba(df_test_all_ct)
    prob_pred = pd.DataFrame(pred, columns= classifier_model.classes_, index=adata.obs.index)
    percent_pred = prob_pred * 100


    ## Find the columns with the largest values
    top_features = percent_pred.idxmax(axis=1)
    ## Get percentage of the top features 
    top_preds = percent_pred.max(axis=1)

    adata.obs['ANNprediction_V1'] = top_features
    adata.obs['ANNprediction_V1_accuracy'] = top_preds.round()
    
   

    check_group = ['Platelet','Stromal']
    adata.obs['ANNprediction_V2'] = adata.obs['ANNprediction_V1']
    adata.obs.ANNprediction_V2[(adata.obs.ANNprediction_V1.isin(check_group)) & (adata.obs.ANNprediction_V1_accuracy < 80)] = 'NA'
    

    adata.obs.ANNprediction_V2[adata.obs.ANNprediction_V2 == 'Mast cell'] = 'Mast cell progenitors'
    
    adata.obs.ANNprediction_V2[adata.obs.ANNprediction_V2 == 'S100A+ preNeutrophil'] = 'change'
    adata.obs.ANNprediction_V2[adata.obs.ANNprediction_V2 == 'S100A+ preNeutrophil (cycling)'] = 'S100A+ preNeutrophil'
    adata.obs.ANNprediction_V2[adata.obs.ANNprediction_V2 == 'change'] = 'S100A+ preNeutrophil (cycling)'

    adata.obs.ANNprediction_V1 = pd.Categorical(adata.obs.ANNprediction_V1, categories = color_palette.keys())
    adata.obs.ANNprediction_V2 = pd.Categorical(adata.obs.ANNprediction_V2, categories = color_palette.keys())

    return adata

```

## Read in your sample here

```python
sample= 'GallenFull_integrated.h5ad'

adata = sc.read_h5ad('test_data/'+sample)
adata
```

<b> Normalization step:</b>
- We normalize over 10000 as it approximate median UMI in AML samples using our current chemistry

```python
#Attention, the data needs to be logNormalized to a count per cell of 10000
# However, this test dataset was alread loaded log-normalized, so we don't do it again

if np.sum(adata.X.astype(np.int32)) == np.sum(adata.X):
    print("Log-Normalizing adata...")
    sc.pp.normalize_per_cell(adata, counts_per_cell_after = 10000)
    sc.pp.log1p(adata)
else:
    if np.all(np.round(np.sum(np.exp(adata.X)-1,axis=1),1) == 10000):
        print("Normalization skipped, the matrix appears to be already normalized.")
    else:
        raise ValueError("Error, the matrix does not contain raw counts and\
                          yet does not appear to be log-normalized to a count per cell of 10000")
```

###  Running the Classification function 

```python
adata = ANNpredictions_V2(adata,features,model)
```
