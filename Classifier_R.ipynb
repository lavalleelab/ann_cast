## Classification 
- Features are taken from HCA reference data (which was run with Cell Ranger (version 5) and  GRCh38 standard reference (Ensembl 98) )

```r
# load features
modeldir='Model/'
features <- read.csv(paste0(modeldir,"ANNClassifier_features.txt"),header=F)
features = as.vector(features)$V1

# load the model from disk
model = reticulate::py_load_object(paste0(modeldir,'ANNClassifier.sav'))
```

```r
ANNpredictions_V2 <- function(seurat_obj,features_all=features,modelfile=modelfile){   
    #seurat_obj was filtered for unexpressed genes. We add genes that are
    #in the model but not in seurat_obj and give them zero expression values.
    # Warning if the same gene name are not used between the classifier and query dataset 
    # this might result in some data getting discarded.
    absent_features <- setdiff(features_all,rownames(seurat_obj))
    length(absent_features)
    temp <- as(matrix(0, ncol = length(absent_features), nrow = length(colnames(seurat_obj))),"sparseMatrix") 
    rownames(temp)=colnames(seurat_obj)
    colnames(temp) <- absent_features

    if(grepl('^5.', Version(seurat_obj))){ 
        #V5
        final_df <- cbind(Matrix::t(seurat_obj[["RNA"]]$data),temp)
    }
    else{
        #V3
        final_df <- cbind(Matrix::t(seurat_obj@assays$RNA@data),temp) 
    }

    # annotate cell types
    df_test_all_ct <- final_df[, features_all]
    pred = modelfile$predict_proba(df_test_all_ct)
    
    percent_pred = data.frame(pred)
    colnames(percent_pred) <- modelfile$classes_
    rownames(percent_pred) <- rownames(seurat_obj@meta.data)
    percent_pred = percent_pred * 100
    

    # ## Find the columns with the largest values
    # Find the maximum value in each row
    max_values <- apply(percent_pred, 1, max)
    # Find the column name corresponding to the maximum value in each row
    max_cols <- apply(percent_pred, 1, function(x) names(percent_pred)[which.max(x)])

    seurat_obj@meta.data['ANNprediction_V1'] = max_cols
    seurat_obj@meta.data['ANNprediction_V1_accuracy'] = max_values


    check_group = c('Platelet','Stromal')
    seurat_obj@meta.data['ANNprediction_V2'] = seurat_obj@meta.data$ANNprediction_V1
    seurat_obj@meta.data$ANNprediction_V2[(seurat_obj@meta.data$ANNprediction_V1 %in% check_group) & (seurat_obj@meta.data$'ANNprediction_V1_accuracy' < 80)] = 'NA'
    

    seurat_obj@meta.data$ANNprediction_V2[seurat_obj@meta.data$ANNprediction_V2 == 'Mast cell'] = 'Mast cell progenitors'
    seurat_obj@meta.data$ANNprediction_V2[seurat_obj@meta.data$ANNprediction_V2 == 'S100A+ preNeutrophil'] = 'S100A+ preNeut'
    seurat_obj@meta.data$ANNprediction_V2[seurat_obj@meta.data$ANNprediction_V2 == 'S100A+ preNeutrophil (cycling)'] = 'S100A+ preNeutrophil'
    seurat_obj@meta.data$ANNprediction_V2[seurat_obj@meta.data$ANNprediction_V2 == 'S100A+ preNeut'] = 'S100A+ preNeutrophil (cycling)'

    seurat_obj@meta.data$ANNprediction_V1 <- factor(seurat_obj@meta.data$ANNprediction_V1, levels=names(color_palette))
    seurat_obj@meta.data$ANNprediction_V2 <- factor(seurat_obj@meta.data$ANNprediction_V2, levels=names(color_palette))
    return (seurat_obj)
}
```

<b> Normalization step:</b>
- We normalize over 10000 as it approximate median UMI in AML samples using our current chemistry

```r
## Normalization in Seurat
## Here, the test dataset was already normalized, but this operation is idempotent in Seurat so do it again to
## demonstrate what is needed when we start from raw count data
GalenFull <- NormalizeData(GalenFull, margin = 1,normalization.method = "LogNormalize", scale.factor = 10000)
```

###  Running the Classification function 

```r
GalenFull <- ANNpredictions_V2(GalenFull,features,model)
```
